```python
messy_data["Age"]=messy_data['Age'].astype(int) #change the datatype of the Age

```

```python
messy_data['Salary ($)']=messy_data['Salary ($)'].str.replace(',','') #remove the , and replace it with nothing 
messy_data['Salary ($)']=messy_data['Salary ($)'].str.rstrip('$') # remove the $
messy_data['Salary ($)']=messy_data['Salary ($)'].astype(float)
print(messy_data['Salary ($)']) #verfying the out_put 
```

```python
messy_data['Email']=messy_data['Email'].str.replace('eve@ex.co','eve@example.com')
messy_data['Email']=messy_data['Email'].str.replace('david@example', 'david@example.com')
messy_data['Email']=messy_data['Email'].str.lower()

```
